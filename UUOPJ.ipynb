# 2. Kneser-Ney metoda

Jedna od najčešće korištenih i najbolje izvedenih metoda za izglađivanje kod modeliranja jezika je Kneser-Ney metoda.
<br>Vjerojatnost pojavljivanja $w_i$ kao nastavka: $$P_{NASTAVAK}(w_i)=\frac{|\{w_{i-1}:c(w_{i-1},w_i)>0\}|}{|\{w_{j-1},w_j:c(w_{j-1},w_j)>0\}|}$$
Vjerojatnost unigrama: $$P_{KN}(w)=\frac{max(c_{KN}(w)-d,0)}{\sum_{w'}{c_{KN}(w')}}$$
Vjerojatnost bigrama: $$P_{KN}(w_i | w_{i-1})=\frac{max(c(w_{i-1}, w_i)-d,0)}{c(w_{i-1})}+\lambda(w_{i-1})P_{NASTAVAK}(w_i)$$
gdje je $$\lambda(w_{i-1})=\frac{d}{c(w_{i-1})}|\{w:c(w_{i-1},w)>0\}|$$
<br>Općenita rekurzivna relacija je $$P_{KN}(w_i | w_{i-n+1}^{i-1})=\frac{max(c_{KN}(w_{i-n+1}^i)-d,0)}{c_{KN}(w_{i-n+1}^{i-1})}+\lambda(w_{i-n+1}^{i-1})P_{KN}(w_i | w_{i-n+2}^{i-1})$$
S tim da je $c_{KN}(*)$ broj $(*)$ za veći red, a broj nastavka $(*)$, to jest broj jedinstvenih sadržaja s jednom riječju od * za manji red. 

## 2.1 Kneser-Ney metoda bez stemiranja

U ovom dijelu sam napisala funkcije za dobivanje vjerojatnosti unigrama, bigrama, trigrama, a i rečenica.

### Vjerojatnost nastavka

```python
def vjerojatnost_nastavka(rijec):
    br=0
    for i in rjecnik_bigrama_treniranje:
        if i.split()[1]==rijec:
            br=br+1
    vjerojatnost=br/len(rjecnik_unigrama_treniranje)
    return vjerojatnost
vjerojatnost_nastavka('the')
```

### Vjerojatnost unigrama

Kako je d fiksan, za unigrame sam koristila d=0.75.

```python
def vjerojatnost_unigrama(rijec):
    if rijec in rjecnik_unigrama_treniranje:      
        vjerojatnost=max(rjecnik_unigrama_treniranje[rijec]-0.75,0)/len(rjecnik_unigrama_treniranje)
    else:
        vjerojatnost=0.75/len(rjecnik_unigrama_treniranje)
    return vjerojatnost
vjerojatnost_unigrama('the')
```

### Vjerojatnost bigrama

Kako je d fiksan, za bigrame sam koristila d=0.75.

```python
def vjerojatnost_bigrama(rijec):
    prva=rijec.split()[0]
    druga=rijec.split()[1]
    br=0
    br1=0
    for i in rjecnik_bigrama_treniranje:
        if i.split()[0]==prva:
            br=br+rjecnik_bigrama_treniranje[i]
            br1=br1+1
    br2=0
    for i in rjecnik_bigrama_treniranje:
        if i.split()[1]==druga:
            br2=br2+1
    if rijec in rjecnik_bigrama_treniranje:      
        vjerojatnost=max(rjecnik_bigrama_treniranje[rijec]-0.75,0)/br+(0.75/br)*br1*(br2/len(rjecnik_bigrama_treniranje))
    else:
        vjerojatnost=0.75/len(rjecnik_bigrama_treniranje)
    return vjerojatnost
vjerojatnost_bigrama('of the')
```

### Vjerojatnost trigrama

U višim redovima n-grama, počevši od trigrama, koristimo d=0 pa je dovoljno izračunati $$\frac{c_{KN}(w_{i-n+1}^i)}{c_{KN}(w_{i-n+1}^{i-1})}$$

```python
def vjerojatnost_trigrama(rijec):
    prva=rijec.split()[0]
    druga=rijec.split()[1]
    treca=rijec.split()[2]
    br=0
    for i in rjecnik_trigrama_treniranje:
        if i.split()[0]==prva and i.split()[1]==druga:
            br=br+rjecnik_trigrama_treniranje[i]
    if rijec in rjecnik_trigrama_treniranje:      
        vjerojatnost=rjecnik_trigrama_treniranje[rijec]/br
    else:
        vjerojatnost=0.75/len(rjecnik_trigrama_treniranje)
    return vjerojatnost
vjerojatnost_trigrama('the chief clerk')
```

### Vjerojatnost rečenice

```python
def vjerojatnost_recenice_unigrami(recenica):
    rijeci=recenica[:-1].split()
    vjerojatnost=1
    for i in range(0, len(rijeci)):
        vjerojatnost=vjerojatnost*vjerojatnost_unigrama(rijeci[i])
    return vjerojatnost
vjerojatnost_recenice_unigrami("The bedding was hardly able to cover it and seemed ready to slide off any moment.")
```

```python
def vjerojatnost_recenice_bigrami(recenica):
    rijeci=recenica[:-1].split()
    vjerojatnost=1
    for i in range(0, len(rijeci)-1):
        vjerojatnost=vjerojatnost*vjerojatnost_bigrama(rijeci[i]+" "+rijeci[i+1])
    return vjerojatnost
vjerojatnost_recenice_bigrami("The bedding was hardly able to cover it and seemed ready to slide off any moment.")
```

```python
def vjerojatnost_recenice_trigrami(recenica):
    rijeci=recenica[:-1].split()
    vjerojatnost=1
    for i in range(0, len(rijeci)-2):
        vjerojatnost=vjerojatnost*vjerojatnost_trigrama(rijeci[i]+" "+rijeci[i+1]+" "+rijeci[i+2])
    return vjerojatnost
vjerojatnost_recenice_trigrami("The bedding was hardly able to cover it and seemed ready to slide off any moment.")
```

# 3. Perpleksija

Perpleksija je vjerojatnost testnog skupa normaliziranog po broju riječi 
$$PP(W)=P(w_1w_2...w_n)^{-\frac{1}{m}}$$
Zbog račala perpleksiju sam radila u logaritamskom prostoru:
$$10^{-\frac{1}{m}\sum_{i=1}^nlog_{10}P(w_i)}$$
W je korpus za testiranje, a $w_i$ je oznaka za i-tu rečenicu, i=1,...,n.
<br>m je oznaka za broj unigrama (bigrama, trigrama) u korpusu za testiranje.

```python
recenice_unigrama_testiranje=[]
recenica=''
for i in range(0,len(korpus_za_testiranje)):
    if (korpus_za_testiranje[i]=='.' and korpus_za_testiranje[i-2:i]!="Mr" and korpus_za_testiranje[i-3:i]!="Mrs") or korpus_za_testiranje[i]=='!' or korpus_za_testiranje[i]=='?':
        recenice_unigrama_testiranje.append(recenica.split())
        recenica=''
    else:
        recenica=recenica+korpus_za_testiranje[i]
```

```python
recenice_bigrama_testiranje=[]
recenica=[]
for i in recenice_unigrama_testiranje:
    for j in range(0,len(i)-1):
        recenica.append(i[j]+" "+i[j+1])
    recenice_bigrama_testiranje.append(recenica)
    recenica=[]
```
